```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# now load the IRIS dataset 
df = sns.load_dataset('iris')
```

```python
# Who are these flowers?
sns.countplot(x='species', data=df)
```

```python
# Are these flowers small or big?  
sns.histplot(df['petal_length'], kde=True,bins=10)
```

```python
# Where are most flowers concentrated?
sns.kdeplot(x='petal_width',data=df, hue='species')
```

```python
# Who is unusual?
sns.boxplot(y='sepal_width',data=df,)
```

```python
# Do features work together?
sns.scatterplot(x='petal_length',y='petal_width',data=df, hue='species')
```

```python
# Average comparison

sns.barplot(y='petal_length', hue='species',data=df)
```

```python
# Distribution inside each species
sns.violinplot(y='petal_length', hue='species',data=df,palette='Set3')
```

```python
# “Which features talk to each other?”
sns.pairplot(df,hue='species')
```

```python
# Relationship strength map”
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm')
```
